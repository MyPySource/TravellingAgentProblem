```python
#import the libraries
import pandas as pd
import numpy as np
```

```python
#import dataset
data = pd.read_excel('QOinput.xlsx')
```

```python
#add location id
data['Location_id'] = np.array(range(1,len(data)+1))
```

```python
#move location id as first col
data = data[['Location_id','Longitude','Latitude','Order Qty (units)']]
```

```python
# sort all geo locations to form closest set data points to each other
# this is a trick beacuse if geo locations are in sorted order then
# sequentially they are close to each other

df = data.sort_values(["Longitude", "Latitude"], ascending = (True, True))
```

```python
df.set_index('Location_id', inplace=True)
```
